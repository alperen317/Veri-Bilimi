### Question 1 (20%)

```python
def get_energy():
    import pandas as pd
    energy= pd.read_excel('Energy Indicators.xls', sheet_name=0, skiprows=16, skip_footer=38, 
                      usecols=['Unnamed: 2', 'Energy Supply',
                               'Energy Supply per capita', 'Renewable Electricity Production'])
    energy.rename(columns={"Unnamed: 2": "Country", "Energy Supply":"Energy Supply" , "Energy Supply per capita": "Energy Supply per Capita",
                      "Renewable Electricity Production":"% Renewable"}, inplace=True)
    energy.drop(0, inplace= True)
    energy['Energy Supply']= energy['Energy Supply']* 1000000

    import numpy as np

    def isnumber(x):
        try:
            float(x)
            return x
        except:
            return np.NaN
    
    energy['Energy Supply']= energy['Energy Supply'].apply(lambda x: isnumber(x))
    energy['Energy Supply per Capita']= energy['Energy Supply per Capita'].apply(lambda x: isnumber(x))
    energy['% Renewable']= energy['% Renewable'].apply(lambda x: isnumber(x))
    energy['Country']= energy['Country'].str.replace('[0-9]','')
    energy['Country']= energy['Country'].str.split('(').str.get(0)
    energy['Country']= energy['Country'].str.strip()

    energy['Country'].replace("Republic of Korea", value= "South Korea", inplace= True)
    energy['Country'].replace("United States of America", value= "United States", inplace= True)
    energy['Country'].replace("United Kingdom of Great Britain and Northern Ireland", value= "United Kingdom", inplace= True)
    energy['Country'].replace("China, Hong Kong Special Administrative Region", value= "Hong Kong", inplace= True)
    energy.dropna(inplace= True)
    energy.set_index('Country', inplace= True)
    return energy
```

```python
def get_GDP():
    import pandas as pd
    GDP = pd.read_csv('world_bank.csv', skiprows=4)
    GDP['Country Name'].replace("Korea, Rep.", value= "South Korea", inplace=True)
    GDP['Country Name'].replace("Iran, Islamic Rep.", value= "Iran", inplace=True)
    GDP['Country Name'].replace("Hong Kong SAR, China", value= "Hong Kong", inplace=True)
    GDP.rename(columns={'Country Name':'Country'}, inplace=True)
    GDP.set_index('Country', inplace=True)
    GDP= GDP[['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
    return GDP
```

```python
def answer_one():
    import pandas as pd
    
    energy= get_energy()
    GDP= get_GDP()
    
    ScimEn= pd.read_excel('scimagojr-3.xlsx')
    ScimEn= ScimEn.where(ScimEn['Rank']<=15).dropna().set_index('Country')

    MergedDF= pd.merge(ScimEn, energy, how='inner', left_index=True, right_index=True) # inner join: intersection
    MergedDF2= pd.merge(MergedDF, GDP, how='inner', left_index=True, right_index=True) # inner join: intersection
    
    return MergedDF2
```

### Question 2 (6.6%)
The previous question joined three datasets then reduced this to just the top 15 entries. When you joined the datasets, but before you reduced this to the top 15 items, how many entries did you lose?

*This function should return a single number.*

```python
def answer_two():
    import pandas as pd
    
    energy= get_energy()
    GDP= get_GDP()
    
    ScimEn= pd.read_excel('scimagojr-3.xlsx')
    ScimEn= ScimEn.set_index('Country')

    MergedDF= pd.merge(ScimEn, energy, how='inner', left_index=True, right_index=True) # inner join: intersection
    MergedDF2= pd.merge(MergedDF, GDP, how='inner', left_index=True, right_index=True) # inner join: intersection
    
    MergedDF_outer= pd.merge(ScimEn, energy, how='outer', left_index=True, right_index=True) # inner join: intersection
    MergedDF2_outer= pd.merge(MergedDF_outer, GDP, how='outer', left_index=True, right_index=True) # inner join: intersection

    return len(MergedDF2_outer)-len(MergedDF2)
```

```python
answer_two()
```
